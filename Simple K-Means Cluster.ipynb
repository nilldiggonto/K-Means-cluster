```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing KMeans from sklearn
```

```python
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('Countries_area.csv')
```

```python
data.head()
```

```python
data
```

```python
#ploting the data 
```

```python
x_axis =  data['Longitude']
y_axis = data['Latitude']
```

```python
plt.scatter(x_axis,y_axis)
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show
```

```python
#choosing only Latitude and longitude
```

```python
x = data.iloc[:,1:]
```

```python
x.head()
```

```python
#cluser with sklearn
```

```python
Kmeans = KMeans(5) #number of cluster that need to be shown
```

```python
Kmeans.fit(x)
```

```python
#finalizing with predict method
```

```python
all_cluster = Kmeans.fit_predict(x)
```

```python
all_cluster
```

```python
#assiging a column to understand their cluster identity
```

```python
new_data = data.copy()
new_data['Cluster'] = all_cluster
new_data.head(20)
```

```python
#plot the data again with color format
```

```python
new_x_axis = new_data['Longitude']
new_y_axis = new_data['Latitude']
```

```python
plt.scatter(new_x_axis,new_y_axis,c=new_data['Cluster'],cmap='rainbow')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show
```
